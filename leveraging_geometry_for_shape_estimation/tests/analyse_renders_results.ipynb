```python
print(data["pixel_diffs_render"][0])
```

```python
diffs = []
for i in range(len(data["pixel_diffs_render"])):
    diffs.append(data["pixel_diffs_render"][i]["n_diff"])
```

```python
import numpy as np
from matplotlib import pyplot as plt
plt.hist(np.array(diffs))
```

```python
import cv2
max_value = max(diffs)
max_index = diffs.index(max_value)

old_path = data["pixel_diffs_render"][max_index]["old_path"]
new_path = data["pixel_diffs_render"][max_index]["new_path"]

old = cv2.imread(old_path)
new = cv2.imread(new_path)

plt.imshow(old)
plt.show()
plt.imshow(new)
plt.show()

mask = (old != new)
mask = np.all(mask,axis=2)

plt.imshow(mask)
plt.show()

```
