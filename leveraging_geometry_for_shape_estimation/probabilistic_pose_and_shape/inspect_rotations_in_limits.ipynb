```python
import os
import json
from scipy.spatial.transform import Rotation as scipy_rot

folder_path = '/scratch/fml35/experiments/leveraging_geometry_for_shape/exp_058_rotation_lines_kp_translation_gt_retrieval_s2/poses_R'

files = []
names = []
for name in os.listdir(folder_path):
    file_path = folder_path + '/' + name
    with open(file_path,'r') as file:
        files.append(json.load(file))
    names.append(name)
```

```python
names_outside_limits = []
for i in range(len(files)):
    if files[i]["in_limits"] == "False":
        rot = list(scipy_rot.from_matrix(files[i]["gt_R"]).as_euler('zyx', degrees=True))
        print(names[i])
        names_outside_limits.append(names[i])
        print('tilt,azim,elev',rot)
```

```python
unique_names = []
for name in names_outside_limits:
    unique_names.append(name.split('_')[0] + '_' + name.split('_')[1])
print(set(unique_names))
print(len(set(unique_names)))
print(sorted(list(set(unique_names))))
```
